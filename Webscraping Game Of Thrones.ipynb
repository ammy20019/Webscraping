```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_episodes'
```

```python
page= requests.get(url)
page.status_code
```

```python
soup = BeautifulSoup(page.text ,'html.parser')
```

```python
episodes_table = soup.find_all('table',class_='wikitable plainrowheaders wikiepisodetable')
```

```python
for table in episodes_table:
    rows=table.find_all('tr')[1:]
    for items in rows:
        season_number=items.find_all('td')[0].get_text()
        director = items.find_all('td')[2].get_text()
        print(season_number,director)
```

```python
episodes = []

for table in episodes_table:
    headers = []
    for header in table.find('tr').find_all('th'):
        headers.append(header.text)
    for row in table.find_all('tr')[1:]:
        values = []
        for col in row.find_all(['th','td']):
            values.append(col.text)
        episode_dict = {headers[i] : values[i] for i in range(len(values))}
        episodes.append(episode_dict)
```

```python
df=pd.DataFrame(episodes)
```
