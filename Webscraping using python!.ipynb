```python
import requests
from bs4 import BeautifulSoup
result = requests.get("https://webscraper.io/test-sites/e-commerce/scroll")
src= result.content
soup = BeautifulSoup(src,'html.parser')
#print(soup.prettify())
```

```python
#print(soup.get_text())
```

```python
urls=[]
div_tag1= soup.find_all('div',class_='thumbnail')

for h in div_tag1:
    div_tag2=h.find_all('div',class_='caption')
    for i in div_tag2:
        try:
            price = i.find_all('h4',class_='pull-right price').get_text()
            desc = i.find_all('p',class_='description').get_text()
        except AttributeError:
            print(price,desc)
        
    
```

```python
products = []

for h in div_tag1:
    div_tag2=h.find_all('div',class_='caption')
    for i in div_tag2:
        price = i.find('h4',class_='pull-right price').get_text()
        desc = i.find('p',class_='description').get_text()
        p_dict = {'name of product':desc,'Price':price}
        products.append(p_dict)
```

```python
import pandas as pd
df = pd.DataFrame(products)
```

```python
import requests
from bs4 import BeautifulSoup
url = requests.get('https://en.wikipedia.org/wiki/Stranger_Things')
soup= BeautifulSoup(url.content,'html.parser')
print(url)
```

```python
episodes_table = soup.find('table',class_='wikitable plainrowheaders wikiepisodetable').get_text()
for table in episodes_table:
    rows=table.find_all('tr',class_='vevent')
    for items in rows:
        season_number=items.find_all('td',class_='text-align:center').get_text()
        director = items.find_all('td',class_='summary').get_text()
        print(season_number,director)
```
